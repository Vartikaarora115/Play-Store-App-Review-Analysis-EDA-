**3.App Size**

We can see that the values in the Size column contains data with different units. 'M' stands for MB and 'k' stands for KB. To easily analyse this column, it is necessary to convert all the values to a single unit. In this case, we will convert all the units to MB.

We know that 1MB = 1024KB, to convert KB to MB, we must divide all the values which are in KB by 1024.

```python
#Cleaning the data for size column:
sizes = [i for i in df['Size']]

def clean_sizes(sizes_list):
    """
    As sizes are represented in 'M' and 'k', we remove 'M'
    and convert 'k'/kilobytes into megabytes
    """
    cleaned_data = []
    for size in sizes_list:
      if 'M' in size:
        size = size.replace('M', '')
        size = float(size)
      elif 'k' in size:
        size = size.replace('k', '')
        size = float(size)
        size = size/1024  # 1 megabyte = 1024 kilobytes
        # representing 'Varies with device' with value 0
      elif 'Varies with device' in size:

        size = float(0)
      cleaned_data.append(size)
    return cleaned_data
```

```python
df['Size']=clean_sizes(sizes)         #size data has been cleaned and converted to float data type
```

```python
df[df['Size'].isnull()] #no null values in size column
```

**4.Installs**

To convert all the values in the Installs column from string datatype to integer datatype, we must first drop the '+' and ',' symbol from all the entries if present and then we can change its datatype.

```python
df['Installs']= df['Installs'].str.replace(',', '',  regex=True)
df['Installs']= df['Installs'].str.replace('+', '',  regex=True)
df['Installs']=df['Installs'].astype(int)
```

```python
df[df['Installs'].isnull()] #no null values in Installs column
```

**5.Price**

To convert the price column from string to float, we must first drop the $ symbol from the all the values. Then we can assign float datatype to those values.

```python
pricelist=[]
def convert_dollar(price_list):
  '''
  This funtion drops the $ symbol if present and returns the value with float datatype.
  '''
  for price in price_list:

    if '$' in price:
      price= (price[1:])
      price=float(price)
    else:
      price= float(price)
    pricelist.append(price)
  return pricelist
```

```python
pricee_list=list(df['Price'])
df['Price']= convert_dollar(pricee_list)
```

# Classification of apps based on different size groups:

```python
# Function to group the apps based on its size in MB
size_list=[]
def app_size(list):

  '''
  This function groups the size of an app 
  between ~0 to 100 MB into certain intervals.
  '''
  for s in list:
    if s > 0 and s<10:
      p= 'Very small'
    elif s >=10 and s <30:
      p= 'Small'
    elif s >= 30 and s <60:
      p= 'Medium'
    elif s >= 60 and s <80:
      p= 'Large'
    elif s >= 80 :
      p= 'Very large'
    
    size_list.append(p)
  return size_list
  
```

```python
df['Size Group']= app_size(list(df['Size']))          #creating a new column for app size group:
```
